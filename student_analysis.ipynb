```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_performance.csv")
df.head()
```

```python
df.isnull().sum()

df.describe()
```

```python
df["average_score"] = df[["math_score", "science_score", "english_score"]].mean(axis=1)
df.head()
```

```python
plt.figure()
sns.scatterplot(data=df, x="study_hours", y="average_score")
plt.title("Study Hours vs Average Student Score")
plt.show()
```

```python
plt.figure()
sns.scatterplot(data=df, x="attendance", y="average_score")
plt.title("Attendance vs Average Student Score")
plt.show()
```

```python
gender_avg = df.groupby("gender")["average_score"].mean()
gender_avg

plt.figure()
gender_avg.plot(kind="bar", title="Average Score by Gender")
plt.show()
```

## Key Insights

- Higher study hours correlate with higher average scores.
- Attendance shows a strong positive relationship with performance.
- Female students show slightly higher average scores in this sample.
- Data supports targeted interventions for low-attendance students.
